```python
import re
import numpy
import math
import matplotlib
import matplotlib.pyplot

def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)

def solve(x):
    n = len(x);
    points = []
    coefs = [[0 for j in range(n)] for i in range(n)]
    
    for i in range(n): 
        points.append(f(x[i]))
        for j in range(n):        
            coefs[i][j] =  x[i] ** j
        
    return(numpy.linalg.solve(coefs, points))

def graph(rank, scope):
    y = numpy.zeros(shape=(len(scope)))
    for i in range(len(scope)):
        for j in range(len(rank)):
                y[i] += rank[j] * (scope[i] ** j)
    return y

steps = numpy.arange(1, 15, 0.5)

w1 = solve([1, 15])
w2 = solve([1, 8, 15])
w3 = solve([1, 4, 10, 15])

a = graph(w1, steps) #1
b = graph(w2, steps) #2
c = graph(w3, steps) #3

matplotlib.pyplot.plot(steps, list(map(f, steps)))
matplotlib.pyplot.plot(steps, a)
matplotlib.pyplot.plot(steps, b)
matplotlib.pyplot.plot(steps, c)
```
